We can derive from the law of Consevation of energy that $$h_{max} = l(1 - \cos\theta_{max})$$ Where $l$ is the distance from the pivit point to the center of gravity of the mass. In physics to relate this problem to the case when we have mass on a string scientists often appoximate the given formula to this one $$h \approx l\frac{\theta^2}{2}$$
We can see that the approximation that they use is $$\cos\theta \approx 1 - \frac{\theta^2}{2}$$
And we can see that this is true for very small values of $\theta$ in radians.


```python
theta = math.radians(0.1)
cos_value = np.cos(theta)
polinomial_value = 1 - theta * theta / 2
print(cos_value)
print(polinomial_value)
print(cos_value - polinomial_value)
```

Now let's graph the two functins.

```python
def plot_function_and_approximation(function, approximation_function, min_x = - 10 , max_x = 10):
    vectorized_function = np.vectorize(function)
    vectorized_approximated_function = np.vectorize(approximation_function)
    
    x = np.linspace(min_x, max_x, 1000)
    y = vectorized_function(x)
    
    x_approximated = np.linspace(min_x, max_x, 1000)
    y_approximated = vectorized_approximated_function(x)
    
    plt.plot(x, y)
    plt.plot(x_approximated, y_approximated)
    plt.grid(True)
    plt.show()
```

```python
plot_function_and_approximation(cos, lambda theta: 1 - theta * theta / 2,- np.pi , np.pi)
```

It is visible that the two functions are practically overlap around zero. 

__ But why and how come do we to this quadratic.__ 

Answer is the study of Taylor siries. The Taylor siries is is a method aiming at representing nonpolinomial functions as polinimials that approximate them near some input. Polinomials are easier to take derivatives of, to compute, to graph, to integrate and so on. <br\>
Let's look at the cosign function again to try and find a polinomial   $P_{(x)} = c_0 + c_1x + c_2x^2$ that approximates $\cos(x)$ near zero, where we choose the constants $c_0,\quad c_1 \quad \textrm{and} \quad c_2 $. <br\> 
First if we want any approximation at all when we input zero our polinomial shold give ус the same value. $\cos(0) = 1$ and if we input in $P_{(x)}$ zero we get $1 = c_0$. So far we have $$ P_{(x)} = 1 + c_2x + c_3x^2$$ 
This gives us the freedome to choose $c_1$ and $c_2$ but what ever we do with them won't change the fact that at zero the polinomial is equal to one.

```python
c_0 = 1
```

```python
plot_function_and_approximation(cos, lambda theta: c_0 + 1 * theta - 2 * theta * theta ,- np.pi / 2 , np.pi / 2)
plot_function_and_approximation(cos, lambda theta: c_0 - 0.5 * theta + 0.1 * theta * theta ,- np.pi * 2 , np.pi * 2)
plot_function_and_approximation(cos, lambda theta: c_0 + 0.2 * theta - 0.2 * theta * theta ,- np.pi * 2, np.pi * 2)
```

As we can see we can choose random $c_1$ and $c_2$ but we need same tanget slope aroud zero. The derivative of $\cos(x)$ is 
$- \sin(x)$ and at zero $- \sin(0) = 0$. $\frac{dP}{dx} = c_1 + 2 c_2  x$. By the same logic we sustitute with zero. So $\frac{dP_{(0)}}{dx} = c_1$. The constant $c_1$ has compleat control over our derivative of our approximation around $x = 0$. Fallowing the same algorithm as with $c_0$ we equalise the the two functions at zero $$- \sin(0) = \frac{dP_{(0)}}{dx}$$ and find that $$0 = c_1 + 2c_2\cdot(0)$$ $$c_1 = 0$$

```python
c_1 = 0
```

```python
plot_function_and_approximation(cos, lambda theta: c_0 + c_1* theta - 0 * theta * theta ,- np.pi * 2 , np.pi * 2)
plot_function_and_approximation(cos, lambda theta: c_0 - c_1 * theta + 0.1 * theta * theta ,- np.pi * 2 , np.pi * 2)
plot_function_and_approximation(cos, lambda theta: c_0 + c_1 * theta - 0.2 * theta * theta ,- np.pi * 2, np.pi * 2)
```

Now the tangent slope around zore in the same as that of $\cos$ and value is the same. As we have seen from the graphs the value of $c_2$ does not change that.<br\> The new step in algorithm is the same: the second derivatative $\cos(x)$ is $- \cos(x)$ ;<br\> $$\frac{d^2P}{dx^2} = 1 \cdot 2  \cdot c_2x^0$$ <br\> $$ -\cos(0) = -1 = 2c_2 =\frac{d^2P_{(0)}}{dx^2} $$ <br\>$$c_2 = -0.5$$ <br\>

With the permission of the author,who named himself "Mic" we will use the code from his  [article](https://firsttimeprogrammer.blogspot.bg/2015/03/taylor-series-with-python-and-sympy.html?showComment=1510074298192#c2495669175052683922)  to calculate and plot the result for $\sin$ in  the Taylor siries.

```python
plt.style.use("ggplot")

# Define the variable and the function to approximate
x = sy.Symbol('x')
f = sin(x)

# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p
```

```python
# Plot results
def plot():
    x_lims = [-5,5]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    # Approximate up until 10 starting from 1 and using steps of 2
    for j in range(1,10,2):
        func = taylor(f,0,j)
        print('Taylor expansion at n='+str(j),func)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label='order '+str(j))
        y1 = []
    # Plot the function to approximate (sine, in this case)
    plt.plot(x1,np.sin(x1),label='sin of x')
    plt.xlim(x_lims)
    plt.ylim([-5,5])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor series approximation')
    plt.show()

plot()
```

### 2.Definition of Taylor siries

By defition the Taylor siries are an infinite siries :
$$\sum_{n=0}^\infty \frac{f^{(n)}(a) \, (x - a)^n}{n!}$$ This is following the algorithm to infinity. The constant $c_n \frac{f^{(n)}(a)}{n!}$ where a is the point at which we are approximating. We substract $a$ from x to cancel it and to get zero.
The Taylor siries around zero are called Maclaurin siries.

### 3. Experiments with the Taylor siries

 It is proven that $\sin, \, cos$ and $  \, e^x$ can be exactly represented by there corresponding siries.Let's see what is the error of the code of $e^x$  if we approximate let's say around -10 with $12^{th}$ degree poinomial. THe value of $e$ to the $50^th$ digit after the decimal point is 2.71828 18284 59045 23536 02874 71352 66249 77572 47093 69995. This number can't be stored as a variable in any computer language but will use it as a pointer to see our accuracy.

```python
def get_value_of_e_to_degree_with_taylor_at_around(at_point, around = 10, degree = 8):
    taylor_polinomial = taylor(exp(x), around, degree)
    print('Value of e to the ' + str(at_point)+ '-th degree  with ' + str(degree) + ' -th degree Taylor polinomial = ',taylor_polinomial.evalf(subs={x:at_point}))
    
get_value_of_e_to_degree_with_taylor_at_around(1, 0, 10)
get_value_of_e_to_degree_with_taylor_at_around(1, 0, 30)
get_value_of_e_to_degree_with_taylor_at_around(1, 0, 49)
```
